```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def load_data(file_name):
    csv_folder = "csvs"
    file_path = f"{csv_folder}/{file_name}.csv"
    data = pd.read_csv(file_path)
    return data
```

```python
def plot_data(data: pd.DataFrame, columns: list):
    for column in columns:
        plt.plot(data[column], label=column)
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.title('Data Plot')
    plt.legend()
    plt.show()
```

```python
data = load_data('bts3v2_palm_down')
data
```

```python
plot_data(data, ['ff_biotac_3'])
plot_data(data, ['slipped'])
```
